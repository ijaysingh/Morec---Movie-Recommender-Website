```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("archive/tmdb_5000_movies.csv")
credits = pd.read_csv("archive/tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits, on = 'title')
movies.info()
```

```python
# genres
# keywords
# title
# overview
# cast
# crew
# id

movies = movies[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
import ast
def convert(obj):
    l = []
    for item in ast.literal_eval(obj):
        l.append(item['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def castFetch(obj):
    count = 0
    l = []
    for item in ast.literal_eval(obj):
        if count != 3:
            l.append(item['name'])
            count += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(castFetch)
movies.iloc[0].cast
```

```python
movies.head()
```

```python
movies['crew'].values
```

```python
def dirFetch(obj):
    l = []
    for item in ast.literal_eval(obj):
        if item['job'] == 'Director':
            l.append(item['name'])
    return l
```

```python
movies['crew'] = movies['crew'].apply(dirFetch)
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[item.replace(" ", "") for item in x])
movies['cast'] = movies['cast'].apply(lambda x:[item.replace(" ", "") for item in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[item.replace(" ", "") for item in x])
movies['crew'] = movies['crew'].apply(lambda x:[item.replace(" ", "") for item in x])
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
movies.head()
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
movies.head()
```

```python
newDf = movies[['movie_id', 'title', 'tags']]
newDf.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for item in text.split():
        y.append(ps.stem(item))
        
    return " ".join(y)
```

```python
newDf['tags'] = newDf['tags'].apply(stem)
```

```python
newDf.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(newDf['tags']).toarray()
```

```python
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movieIndex = newDf[newDf['title'] == movie].index[0]
    distances = similarity[movieIndex]
    moviesList = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for item in moviesList:
        print(newDf.iloc[item[0]].title)
```

```python
recommend('Spider-Man')
```
